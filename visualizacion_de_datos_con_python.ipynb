# Visualizacion de datos con Python: Matplotlib, Seaborn y Plotly

```python
import pandas as pd

datos = pd.read_csv('/content/inmigrantes_canada.csv')
```

```python
datos.head()
```

```python
datos.shape
```

```python
datos.info()
```

```python
datos.set_index('Pais',inplace = True)
```

```python
anos = list(map(str,range(1980,2014)))
anos
```

```python
colombia = datos.loc['Colombia',anos]
colombia
```

```python
col_dict = {'Año':colombia.index.tolist(),
            'Inmigrantes':colombia.values.tolist()}

col_dict
```

```python
datos_col = pd.DataFrame(col_dict)
datos_col.tail()
```

```python
 import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8,4))
plt.plot(datos_col['Año'],datos_col['Inmigrantes'])
plt.xticks(['1980','1985','1990','1995','2000','2005','2010'])
plt.title('Inmigracion de Colombianos hacia Canada')
plt.xlabel('Año')
plt.ylabel('Numero de Inmigrantes')
plt.show()
```

# Creando una figura

```python
fig, ax = plt.subplots(figsize=(8,4))
ax.plot(datos_col['Año'],datos_col['Inmigrantes'])
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
ax.set_title('Inmigracion de Colombianos hacia Canada\n1980 - 2013')
ax.set_xlabel('Año')
ax.set_ylabel('Numero de Inmigrantes')
plt.show()
```

```python
fig, axs = plt.subplots(1,2,figsize=(10,3))

axs[0].plot(datos_col['Año'],datos_col['Inmigrantes'])
axs[0].set_title('Inmigracion de colombianos hacia Canada\n1980 - 2013')
axs[0].set_xlabel('Año')
axs[0].set_ylabel('Numero de Inmigrantes')
axs[0].xaxis.set_major_locator(plt.MultipleLocator(5))
axs[0].grid()

axs[1].boxplot(datos_col['Inmigrantes'])
axs[1].set_title('Boxplot de la inmigracion de colombianos\nhacia Canada 1980 - 2013')
axs[1].set_xlabel('Colombia')
axs[1].set_ylabel('Numero de Inmigrantes')
axs[1].grid()

plt.show()
```

```python
fig, axs = plt.subplots(2,2,figsize=(10,6))
fig.subplots_adjust(hspace=0.5,wspace=0.3)

axs[0,0].plot(datos.loc['Colombia',anos])
axs[0,0].set_title('Colombia')

axs[0,1].plot(datos.loc['Brasil',anos])
axs[0,1].set_title('Brasil')

axs[1,0].plot(datos.loc['Argentina',anos])
axs[1,0].set_title('Argentina')

axs[1,1].plot(datos.loc['Perú',anos])
axs[1,1].set_title('Perú')

for ax in axs.flat:
  ax.xaxis.set_major_locator(plt.MultipleLocator(5))
  ax.set_xlabel('Año')
  ax.set_ylabel('Numero de Inmigrantes')
  ax.grid()

plt.show()
```

```python
#Establecer escala
fig, axs = plt.subplots(2,2,figsize=(10,6))
fig.subplots_adjust(hspace=0.5,wspace=0.3)
fig.suptitle('Inmigracion de los 4 mayores paises sudamericanos\nhacia Canada desde 1980 hasta 2013')

axs[0,0].plot(datos.loc['Colombia',anos])
axs[0,0].set_title('Colombia')

axs[0,1].plot(datos.loc['Brasil',anos])
axs[0,1].set_title('Brasil')

axs[1,0].plot(datos.loc['Argentina',anos])
axs[1,0].set_title('Argentina')

axs[1,1].plot(datos.loc['Perú',anos])
axs[1,1].set_title('Perú')

ymin = 0
ymax = 7000

for ax in axs.ravel():
  ax.set_ylim(ymin,ymax)


for ax in axs.flat:
  ax.xaxis.set_major_locator(plt.MultipleLocator(5))
  ax.set_xlabel('Año')
  ax.set_ylabel('Numero de Inmigrantes')
  ax.grid()

plt.show()
```

```python
colores = []
for pais in sudamerica_sorted.index:
  if pais == 'Colombia':
    colores.append('forestgreen')
  else:
    colores.append('silver')

fig, ax = plt.subplots(figsize=(10,3))

ax.set_frame_on(False)
ax.xaxis.set_visible(False)
ax.tick_params(axis='y',which='both',size=0)

ax.barh(sudamerica_sorted.index, sudamerica_sorted['Total'], color= colores)
ax.set_title('Inmigracion de colombianos hacia Canada\ndurante el periodo de 1980 hasta 2013',
             loc='left',fontsize=16)
ax.set_ylabel('')
ax.yaxis.set_tick_params(labelsize=10)
#ax.set_xlabel('Numero de Inmigrantes',fontsize=14)
#ax.xaxis.set_tick_params(labelsize=12)

for i,j in enumerate(sudamerica_sorted['Total']):
  ax.text(j+20,i,str(j),color='black',fontsize=8.5,ha='left',va='center')

fig.savefig('/content/inmigrantes_sudamericanos_canada.png',
            transparent=False,dpi=300,bbox_inches='tight')

plt.show()
```

```python
ax = sns.barplot(data=top_10,x='Total',y=top_10.index,orient='h')
ax.set(title='Top 10 paises con  mayor migracion hacia Canada\n1980 - 2013',
       xlabel='Numero de Inmigrantes',
       ylabel='Año')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8,4))

ax = sns.barplot(data=top_10,x='Total',y=top_10.index,orient='h')
ax.set_title('Top 10 paises con mayor migracion hacia Canada\n1980 - 2013',
             loc='left',fontsize=16)
ax.set_xlabel('Numero de Inmigrantes',fontsize=14)
ax.set_ylabel('')
plt.show()
```

```python
def generar_grafico(palette):
  fig, ax = plt.subplots(figsize=(8,4))
  ax.set_frame_on(False)
  #ax.xaxis.set_visible(False)
  #ax.tick_params(axis='y',which='both',size=0)
  ax = sns.barplot(data=top_10,x='Total',y=top_10.index,hue=top_10.index,orient='h',
                   palette=palette,legend=False)
  ax.set_title('Top 10 paises con mayor migracion hacia Canada\n1980 - 2013',
              loc='left',fontsize=16)
  ax.set_xlabel('Numero de Inmigrantes',fontsize=14)
  ax.set_ylabel('')
  for i,j in enumerate(top_10['Total']):
    ax.text(j+20,i,str(j),color='black',fontsize=8.5,ha='left',va='center')
  plt.show()
```

```python
def generar_grafico(palette):
  fig, ax = plt.subplots(figsize=(8,4))
  ax.set(xticklabels=[])
  ax = sns.barplot(data=top_10,x='Total',y=top_10.index,hue=top_10.index,orient='h',
                   palette=palette,legend=False)
  ax.set_title('Top 10 paises con mayor migracion hacia Canada\n1980 - 2013',
              loc='left',fontsize=16)
  ax.set_xlabel('Numero de Inmigrantes',fontsize=14)
  ax.set_ylabel('')
  sns.despine()
  for i,j in enumerate(top_10['Total']):
    ax.text(j+20,i,str(j),color='black',fontsize=8.5,ha='left',va='center')
  plt.show()
```

# Grafica Lapso de tiempo

```python
datos_col['Año'] = datos_col['Año'].astype(int)
import plotly.graph_objs as go
```

```python
fig = go.Figure()
```

```python
fig.add_trace(
    go.Scatter(x=[datos_col['Año'].iloc[0]], y=[datos_col['Inmigrantes'].iloc[0]], mode='lines', name='Inmigrantes', line=dict(width=4))
)
```

```python
fig.update_layout(
    title=dict(
        text='<b>Inmigración de Colombianos hacia Canadá en el periodo de 1980 a 2013</b>',
        x=0.12,
        xanchor='left',
        font=dict(size=20)
    ),
    xaxis=dict(range=[1980, 2013], autorange=False, title='<b>Año</b>'),
    yaxis=dict(range=[0, 7000], autorange=False, title='<b>Número de Inmigrantes</b>'),
)
```

```python
fig.update_layout(
    updatemenus=[
        dict(
            type='buttons',
            showactive=False,
            buttons=[
                dict(
                    label='Play',
                    method='animate',
                    args=[None, {
                        'frame': {'duration': 100, 'redraw': True},
                        'fromcurrent': True
                    }]
                )
            ]
        )
    ],
    width=1000,
    height=500
)
```

```python
frames = [go.Frame(data=[go.Scatter(x=datos_col['Año'].iloc[:i+1], y=datos_col['Inmigrantes'].iloc[:i+1])]) for i in range(len(datos_col))]
fig.frames = frames
```

```python
import plotly.graph_objs as go

# Criando uma figura
fig = go.Figure()

# Adicionando a linha do gráfico e definindo a espessura da linha
fig.add_trace(
    go.Scatter(x=[datos_col['Año'].iloc[0]], y=[datos_col['Inmigrantes'].iloc[0]], mode='lines', name='Inmigrantes', line=dict(width=4))
)

# Definir la configuración del layout
fig.update_layout(
    title=dict(
        text='<b>Inmigración de Colombianos hacia Canadá en el periodo de 1980 a 2013</b>',
        x=0.12,
        xanchor='left',
        font=dict(size=20)
    ),
    xaxis=dict(range=[1980, 2013], autorange=False, title='<b>Año</b>'),
    yaxis=dict(range=[0, 7000], autorange=False, title='<b>Número de Inmigrantes</b>'),
    updatemenus=[dict(
        type='buttons',
        showactive=False,
        buttons=[dict(
            label='Play',
            method='animate',
            args=[None, {'frame': {'duration': 100, 'redraw': True}, 'fromcurrent': True}]
        )]
    )],
    width=1000,
    height=500
)

# Definir la configuración de la animación
frames = [go.Frame(data=[go.Scatter(x=datos_col['Año'].iloc[:i+1], y=datos_col['Inmigrantes'].iloc[:i+1])]) for i in range(len(datos_col))]
fig.frames = frames

# Mostrar la figura
fig.show()
```
